# 03 Model Testing

This notebook trains the first baseline model for the NFL forecasting project.

The goal is to use the pre-game features created in the feature engineering notebook to predict whether the home team wins each game.

```python
# Import packages
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Load the modeling dataset
modeling_data = pd.read_csv("../data/processed/modeling_dataset_2023_2025.csv")

modeling_data.head()
```

```python
# Check the data
modeling_data.shape
modeling_data.columns.tolist()
modeling_data[["season", "week", "home_team", "away_team", "home_score", "away_score", "home_team_won"]].head()
```

```python
# Choose model features
features = [
    "avg_points_scored_diff",
    "avg_points_allowed_diff",
    "avg_point_diff_diff",
    "win_pct_diff"
]

target = "home_team_won"
```

```python
# Create X and y
X = modeling_data[features]
y = modeling_data[target]
X.head()
y.head()
```

```python
# Check for missing values
X.isna().sum()
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

print("Training rows:", X_train.shape[0])
print("Testing rows:", X_test.shape[0])
```

```python
# Train the logistic regression model
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(X_test)
```

```python
# Check model accurracy
accuracy = accuracy_score(y_test, y_pred)

print(f"Model accuracy: {accuracy:.2%}")
```

```python
# Compare to a simple baseline
home_team_baseline = [1] * len(y_test)

baseline_accuracy = accuracy_score(y_test, home_team_baseline)

print(f"Baseline accuracy if always picking home team: {baseline_accuracy:.2%}")
```

```python
# Confusion matrix
confusion_matrix(y_test, y_pred)
```

The confusion matrix shows how many games the model predicted correctly and incorrectly.

Rows are the actual results.  
Columns are the predicted results.

```python
# Classification report
print(classification_report(y_test, y_pred))
```

```python
# View model coefficients
coefficients = pd.DataFrame({
    "feature": features,
    "coefficient": model.coef_[0]
})

coefficients.sort_values("coefficient", ascending=False)
```

```python
# Create predicted probabilities
y_prob = model.predict_proba(X_test)[:, 1]

y_prob[:10]
```

```python
# Create a results table
results = modeling_data.loc[X_test.index, [
    "season",
    "week",
    "home_team",
    "away_team",
    "home_score",
    "away_score",
    "home_team_won"
]].copy()

results["predicted_home_win"] = y_pred
results["home_win_probability"] = y_prob

results.head(10)
```

```python
# Add predicted winner
results["predicted_winner"] = results.apply(
    lambda row: row["home_team"] if row["predicted_home_win"] == 1 else row["away_team"],
    axis=1
)

results["actual_winner"] = results.apply(
    lambda row: row["home_team"] if row["home_team_won"] == 1 else row["away_team"],
    axis=1
)

results.head(10)
```

```python
# Add correctness column
results["correct_prediction"] = (
    results["predicted_winner"] == results["actual_winner"]
)

results.head(10)
```

```python
# Look at correct predictions
results[results["correct_prediction"] == True].head(10)
```

```python
# Look at incorrect predictions
results[results["correct_prediction"] == False].head(10)
```

```python
# Save test predictions
results.to_csv("../data/predictions/baseline_model_test_predictions.csv", index=False)

print("Saved baseline model test predictions.")
print(results.shape)
```

```python
# Final summary numbers
print("Model accuracy:", f"{accuracy:.2%}")
print("Home-team baseline accuracy:", f"{baseline_accuracy:.2%}")
print("Number of test games:", len(results))
print("Correct predictions:", results["correct_prediction"].sum())
print("Incorrect predictions:", len(results) - results["correct_prediction"].sum())
```

In this notebook, I trained the first baseline model for the NFL forecasting project.
