```python
# Import packages
import pandas as pd
```

```python
# Load the schedule data
schedules = pd.read_csv("../data/raw/schedules_2023_2025.csv")
```

```python
# Preview the data
schedules.head()
```

```python
# Check dataset size
schedules.shape
```

```python
# View all column names
schedules.columns.tolist()
```

```python
# View basic game information
schedules[
    ["season", "week", "home_team", "away_team", "home_score", "away_score"]
].head(10)
```

```python
# Check missing scores
schedules[["season", "week", "home_team", "away_team", "home_score", "away_score"]].isna().sum()
```

```python
# Keep only completed games
completed_games = schedules.dropna(subset=["home_score", "away_score"]).copy()\
# Then check size
completed_games.shape
```

```python
# Create the target variable
completed_games["home_team_won"] = (
    completed_games["home_score"] > completed_games["away_score"]
).astype(int)

# Preview it
completed_games[
    ["season", "week", "home_team", "away_team", "home_score", "away_score", "home_team_won"]
].head(10)
```

```python
# Check home win rate
home_win_rate = completed_games["home_team_won"].mean()

home_win_rate
```

```python
# Create a clean game results dataset
game_results = completed_games[
    [
        "season",
        "week",
        "game_id",
        "gameday",
        "home_team",
        "away_team",
        "home_score",
        "away_score",
        "home_team_won"
    ]
].copy()

# Preview
game_results.head()
```

```python
# Add point differential
game_results["home_point_diff"] = (
    game_results["home_score"] - game_results["away_score"]
)

# Preview
game_results.head()
```

```python
# Save processed dataset
game_results.to_csv("../data/processed/game_results_2023_2025.csv", index=False)

# Confirm it saved
print("Saved processed game results dataset.")
print(game_results.shape)
```

```python
# Quick summary
print("Seasons included:", game_results["season"].unique())
print("Total completed games:", len(game_results))
print("Home win rate:", f"{game_results['home_team_won'].mean():.2%}")
print("Average home point differential:", round(game_results["home_point_diff"].mean(), 2))
```
